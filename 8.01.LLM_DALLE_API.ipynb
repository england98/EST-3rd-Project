```python
import openai
import requests
from PIL import Image
from io import BytesIO
```

```python
# OpenAI API 키 설정
openai.api_key = "YOUR_API_KEY" # 생성한 키 정보 입력
```

```python
# 텍스트 프롬프트 정의
prompt = "A futuristic cityscape with flying cars in the style of cyberpunk"
```

```python
# 이미지 생성 요청
response = openai.Image.create(
    prompt=prompt,
    n=1,  # 생성할 이미지 수
    size="512x512"  # 이미지 크기
)
```

```python
# 생성된 이미지 URL 가져오기
image_url = response['data'][0]['url']

# 이미지 다운로드 및 표시
response = requests.get(image_url)
# img = Image.open(BytesIO(response.content))
# img.show()  # 로컬 환경에서 이미지 팝업
```
